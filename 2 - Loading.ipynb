## Create Staging Tables

### Customers

```python
query = '''

CREATE TABLE customers (
    stage_id serial,
    sale_id char(10),
    customer_id char(7),
    first_name varchar(25),
    last_name varchar(25),
    street varchar(50),
    city varchar(25),
    state char(2),
    zip char(5)
)

'''

connection.rollback()
cursor.execute(query)
connection.commit()
```

### Sales

```python
query = '''

CREATE TABLE sales (
    stage_id serial,
    sale_id char(10),
    sale_date char(10),
    sub_total varchar(10),
    tax varchar(10),
    total varchar(10)
)

'''

connection.rollback()
cursor.execute(query)
connection.commit()
```

### Stores

```python
query = '''

CREATE TABLE stores (
    stage_id serial,
    sale_id char(10),
    store_id char(5),
    name varchar(25),
    street varchar(50),
    city varchar(25),
    state char(2),
    zip char(5)
)

'''

connection.rollback()
cursor.execute(query)
connection.commit()
```

### Products

```python
query = '''

CREATE TABLE products (
    stage_id serial,
    sale_id char(10),
    item_id varchar(10),
    product_id char(10),
    price varchar(10),
    quantity varchar(10)
)

'''

connection.rollback()
cursor.execute(query)
connection.commit()
```

## Load

```python
query = '''

COPY customers (sale_id, customer_id, first_name, last_name, street, city, state, zip)
FROM '/Users/bryguan/demo/customers.csv' DELIMITER ',' NULL '' CSV HEADER;

COPY sales (sale_id, sale_date, sub_total, tax, total)
FROM '/Users/bryguan/demo/sales.csv' DELIMITER ',' NULL '' CSV HEADER;

COPY stores (sale_id, store_id, name, street, city, state, zip)
FROM '/Users/bryguan/demo/stores.csv' DELIMITER ',' NULL '' CSV HEADER;

COPY products (sale_id, item_id, product_id, price, quantity)
FROM '/Users/bryguan/demo/products.csv' DELIMITER ',' NULL '' CSV HEADER;

'''

connection.rollback()
cursor.execute(query)
connection.commit()
```
